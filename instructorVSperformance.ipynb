```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('disliking your instructor and grade performance.csv')
df = pd.DataFrame(data)
df3 = data.groupby(['What is your faculty?']) ['What is your faculty?'].count().reset_index(name='count')
df3
```

```python
fig = plt.figure(figsize =(10, 7))
plt.pie(df3['count'], labels=df3['What is your faculty?'],autopct='%1.1f%%' )
plt.title ('What is your faculty?')
plt.show()
```
